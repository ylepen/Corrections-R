## Question 1: Generation d'un bruit blanc

$$x_{t} = \epsilon_{t}\sim NID(0,1)$$

### Réponses:



> Simulation d'un processus bruit blanc





```r
n=1000
e = rnorm(n = n, 0,1)
```



> Graphique de la trajectoire du processus



```r
plot(x = e, type="l",col="blue",main = "Bruit blanc gaussien",ylab = "")
```

> Autocorrelogramme d'un bruit blanc

```r
acf(e, main="Autocorrelogramme d'un bruit blanc")
```

> Autocorrelation partielle d'un bruit blanc

```r
pacf(e, main ="Autocorrelation partielle d'un bruit blanc")
```

### Question 1.a: Generation d'une marche aleatoire sans derive




1.   Option 1 : une boucle

```r
rw1 = rep(0,n)

for(i in 2:n ){
  rw1[i]=rw1[i-1]+e[i]
}
```

```r
plot(x = rw1, type="l",col="blue",main = "Marche aléatoire ",ylab = "")
```

2.   Utilisation de la fonction cumsum

```r
rw2 =cumsum(e[2:n])
```

```r
plot(x = rw2, type="l",col="blue",main = "Marche aléatoire ",ylab = "")
```

```r
acf(rw2)
```

```r
pacf(rw2)
```

### Question 1.b: Generation d'une marche aléatoire avec dérive








1.   Avec constante

```r
mu=1
rw = cumsum(e+mu)
```

```r
plot(x = rw, type="l",col="blue",main = "Marche aléatoire ",ylab = "")
```

```r
acf(rw)
```

```r
pacf(rw)
```

2.   Avec tendance temporelle (time trend)

```r
time_trend = 1:n
```

```r
plot(time_trend,type ="l")
```

```r
a =1
b=0.2
TS = a + b*time_trend+10*e
```

```r
plot(TS,type ="l",col="red")
lines(a+b*time_trend,col="blue")
```

```r
acf(TS)
```

```r
pacf(TS)
```

## Question 2: Generation d'un processus AR(1)

 $$(1-0.8L)x_{t} = \epsilon_{t}$$

###Réponses:

> Simulation

```r
phi1 = 0.9
x1 = rep(0,n)
for(i in 2:n){
x1[i]=phi1*x1[i-1]+e[i]
}
```

> Trajectoire

```r
plot(x = x1, type="l",col="blue",main = "AR(1)",ylab = "")
```

> Autocorrelogramme

```r
acf(x1)
```

> Autocorrelogramme partiel

```r
pacf(x1)
```

## Question 5: Generation d'un processus MA(1)

$$x_{t} = (1 - 0.8 L) \epsilon_{t}$$

###Réponses:

> Simulation

```r
theta1 = 0.9
x2 = rep(0,n)
for(i in 2:n){
  x2[i]=theta1*e[i-1]+e[i]
}
```

> Trajectoire

```r
plot(x = x2, type="l",col="blue",main = "MA(1)",ylab = "")
```

> Autocorrelogramme

```r
acf(x2)
```

> Autocorrelogramme partiel

```r
pacf(x2)
```
